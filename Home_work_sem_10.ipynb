```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmi':lst})
data.head()
```

```python
pd.get_dummies(data['whoAmi'])
```

```python
data.loc[data['whoAmi'] == 'human', 'human'] = '1'
data.loc[data['whoAmi'] == 'robot', 'human'] = '0'
data.loc[data['whoAmi'] == 'human', 'robot'] = '0'
data.loc[data['whoAmi'] == 'robot', 'robot'] = '1'
```

```python
data
```

```python
def my_get_dummies(data, prefix=None, prefix_sep='_', dummy_na=False, columns=None, drop_first=False):
  if columns is not None:
        data = data[columns]
    
   
  dummy_cols = pd.get_dummies(data, prefix=prefix, prefix_sep=prefix_sep, dummy_na=dummy_na, drop_first=drop_first)
    
  return dummy_cols
```

```python
my_get_dummies(data)
```
